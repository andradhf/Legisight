##Chunking text server

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = PyPDFLoader("putusan_mkri.pdf")
documents = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)
chunks = text_splitter.split_documents(documents)

print(f"Total potongan teks: {len(chunks)}")
```

```python
!ollama pull nomic-embed-text
from langchain.embeddings import OllamaEmbeddings
from langchain.vectorstores import FAISS

embedding = OllamaEmbeddings(base_url="http://localhost:11434",model="nomic-embed-text")
vectorstore = FAISS.from_documents(chunks, embedding)
retriever = vectorstore.as_retriever()
```

```python
from langchain.llms import Ollama

llm = Ollama(
    base_url="http://localhost:11434",
    model="deepseek-r1:7b"
)
```

```python
from langchain.chains import RetrievalQA

qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    retriever=retriever,
    return_source_documents=False
)
```

```python
question = ( """
Berdasarkan Context yang di sediakan Tugas Anda adalah menentukan kebjikana atau putusan-putusan tersebt ke dalam komisi-komisi DPR RI berdasarkan ruang lingkup tugas mereka. Gunakan informasi berikut sebagai acuan:

Komisi I: Pertahanan, Luar Negeri, Komunikasi dan Informatika, Intelijen
Komisi II: Pemerintah Dalam Negeri, Pertanahan, Pemberdayaan aparatur
Komisi III: Penegakan Hukum
Komisi IV: Pertanian, Kehutanan, Kelautan
Komisi V: Infrastruktur, Perhubungan
Komisi VI: Perdagangan, BUMN, Persaingan Usaha
Komisi VII: Perindustrian, UMKM, Ekonomi Kreatif, Pariwisata
Komisi VIII: Agama, Sosial, Pemberdayaan Perempuan dan Perlindungan anak
Komisi IX: Kesehatan, Ketenagakerjaan, Jaminan sosial
Komisi X: Pendidikan, Olahraga, Sains, Teknologi
Komisi XI: Keuangan, Moneter, Sektor Jasa Keuangan, Pembangunan Nasional
Komisi XII: Energi dan Sumber Daya Mineral (ESDM), Lingkungan Hidup, Investasi
Komisi XIII: Reformasi regulasi, Hak Asasi Manusia

Penting:
- Jangan ubah struktur atau isi kalimat.
- Pertahankan semua kata sambung seperti *dan*, *atau*, *jika*.
- Jika sebuah kalimat tidak masuk dalam ruang lingkup komisi manapun, abaikan.
- Hanya kelompokkan kalimat yang memiliki perintah, rencana, kebijakan, atau keputusan yang dapat dikaitkan dengan komisi tersebut.

    """
)

answer = qa_chain.run(question)
print("Jawaban:", answer)
```
